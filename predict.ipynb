```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
escuela= pd.read_csv('dataset.csv')
```

```python
escuela.head()
```

```python
escuela.describe()
```

```python
escuela.info('display.max_info_columns')
```

```python
escuela.columns
```

```python
escuela['Target_numeric'] = escuela['Target'].apply(lambda x: 1 if x == 'Graduate' else 0)
```

```python
correlaciones = escuela.corr(numeric_only=True)['Target_numeric'].sort_values(ascending=False)
correlaciones
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(escuela.corr(numeric_only=True).round(2), annot=True, fmt=".1f", cmap='coolwarm', linewidths=0.5)
plt.title('Mapa de calor de correlaciones ')
plt.show()
```
